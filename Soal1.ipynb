```python
24# Array kosong, akan diisi dengan input dari angka = int(input("Masukan Angka"))
arr = []

# Perulangan Pemngimputan Angka sebanyak 10
for i in range(10):
  angka = int(input("Masukan Angka  : "))
  arr.append(angka)

# Penampilan Data (Seluruh Array)
print("Array:", arr)

# Penampilan Data (Nilai Tertinggi)
print("Nilai Tertinggi:", max(arr))

#Penampilan Data (Nilai Terkecil)
print("Nilai Terkecil:", min(arr))

#Penampilan Data (Nilai Rata-rata)
print("Rata-rata:", sum(arr)/len(arr))
```

```python
def insert_tengah(arr, posisi, nilai):
    # tambah ruang di akhir
    arr.append(0)

    # geser ke kanan
    for i in range(len(arr) - 1, posisi, -1):
        arr[i] = arr[i - 1]

    # isi nilai baru
    arr[posisi] = nilai

insert_tengah(arr, 1, 10)
print("Array Setelah Insert Tengah")
print(arr)

def delete_tengah(arr, posisi):
    # geser ke kiri
    for i in range(posisi, len(arr) - 1):
        arr[i] = arr[i + 1]

    # potong elemen terakhir
    arr[:] = arr[:-1]

delete_tengah(arr, 1)
print("Array Setelah Delete Tengah")
print(arr)
```
