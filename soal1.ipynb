```python
24# Array kosong, akan diisi dengan input dari angka = int(input("Masukan Angka"))
arr = []

# Perulangan Pemngimputan Angka sebanyak 10
for i in range(10):
  angka = int(input("Masukan Angka  : "))
  arr.append(angka)

# Penampilan Data (Seluruh Array)
print("Array:", arr)

# Penampilan Data (Nilai Tertinggi)
print("Nilai Tertinggi:", max(arr))

#Penampilan Data (Nilai Terkecil)
print("Nilai Terkecil:", min(arr))

#Penampilan Data (Nilai Rata-rata)
print("Rata-rata:", sum(arr)/len(arr))


```
